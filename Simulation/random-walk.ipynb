```python
import random 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def generate_rw(n, origin):
    n = 100
    origin = 0
    walks = [origin]

    for i in range(n):
        n_step = random.choice([1,-1])
        walks.append(walks[i] + n_step)
    
    return walks

realisations = pd.DataFrame()
w = 50
n = 250
origin = 0 

for j in range(w):
    realisations[str(j)] =  generate_rw(n, origin)

realisations = realisations.stack().reset_index().drop(columns="level_0").rename(columns={"level_1" : "w", 0: "value"})
sns.lineplot(data= realisations.reset_index(), x="index", y="value", hue="w", legend = False)
```
